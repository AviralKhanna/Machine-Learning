```python
from googletrans import Translator

def translate_transcript(transcript_path):
    # Read the text from the transcript file
    with open(transcript_path, 'r', encoding='utf-8') as file:
        lines = file.readlines()

    # Initialize the translator
    translator = Translator()

    # List of words to exclude from translation
    exclude_words = ['sir', 'kya']  # Add more words as needed

    # Translate Urdu to Hindi and English, excluding specific words
    hindi_lines = []
    english_lines = []
    for line in lines:
        hindi_line = ' '.join(translator.translate(word, src='ur', dest='hi').text if word not in exclude_words else word for word in line.split())
        english_line = ' '.join(translator.translate(word, src='ur', dest='en').text if word not in exclude_words else word for word in line.split())
        hindi_lines.append(hindi_line)
        english_lines.append(english_line)

    # Create a new text file for the translated transcript
    output_path = transcript_path.replace('.txt', '_translated.txt')
    with open(output_path, 'w', encoding='utf-8') as output_file:
        for urdu, hindi, english in zip(lines, hindi_lines, english_lines):
            output_file.write(f'Urdu: {urdu.strip()}\n')
            output_file.write(f'Hindi: {hindi.strip()}\n')
            output_file.write(f'English: {english.strip()}\n\n')

    print(f'Translated transcript saved to {output_path}')

if __name__ == "__main__":
    # Replace 'your_transcript.txt' with the actual path to your Urdu transcript text file
    transcript_path = '/content/drive/MyDrive/Autistic/Abnormal_transript/A004_transcript.txt'
    translate_transcript(transcript_path)
```
